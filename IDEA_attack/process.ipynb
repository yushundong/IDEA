```python
import numpy as np

# Given joint distribution of X and Y
joint_distribution = np.array([[3/10, 1/10], [1/10, 2/10], [1/10, 2/10]])

# Calculate marginal probability of X
p_x_0 = joint_distribution[:, 0].sum()
p_x_1 = joint_distribution[:, 1].sum()

# Calculate conditional PMF of Y given X = 0 and X = 1
conditional_pmf_x_0 = joint_distribution[:, 0] / p_x_0
conditional_pmf_x_1 = joint_distribution[:, 1] / p_x_1

# Calculate the expected value of Y given X = 0 and X = 1 (f**)
f_star_star_0 = sum(y * prob for y, prob in enumerate(conditional_pmf_x_0))
f_star_star_1 = sum(y * prob for y, prob in enumerate(conditional_pmf_x_1))

# Calculate expected risk of f**
expected_risk = [
    (y - (f_star_star_0 if x == 0 else f_star_star_1))**2 * p_xy
    for x in range(2)
    for y, p_xy in enumerate(joint_distribution[:, x])
]

conditional_pmf_x_0, conditional_pmf_x_1, f_star_star_0, f_star_star_1, expected_risk
```
